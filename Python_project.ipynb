# <font color="springgreen">**Disabled Population, 2011 - INDIA**</font>

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/content/drive/MyDrive/1_Disabled Population by type of Disability, type of Households and Sex, 2011 - INDIA.csv")
df.head(10)
```

```python
df.shape
```

# <font color="springgreen">**Q1. Total persons with disabilities in both rural and urban area of each state ?**</font>

```python
Total = df[(df["Type of Household"]=="Total") & (df["Area Name"]!="INDIA")]
Total.sort_values(by=["Area Name","Total disabled population - Persons"],ascending=False)
```

# <font color="orchid">**Insights from above :**</font>


*   West Bengal has the highest number(1589266) of persons with disabilities, of which 1107498 are in rural areas and the rest(481768) in urban areas.
*   Andaman & Nicobar Islands has the lowest number(5702) of persons with disabilities, of which 4152 are in rural areas and the rest(1550) in urban areas.


*   There are more male(1127181) with disabilities in West Bengal than female(890225).
*   In Andabar and Nicobar, there are more female(201) suffering from mental illness than male(163).





# <font color="springgreen">**Q2. What are the 'Houseless Household' population in India, including rural and urban areas?**</font>


```python
df1 = df[(df["Type of Household"]=="Houseless Household") & (df["Area Name"]=="INDIA")]
df1
```

```python
df1.groupby(['Total/ Rural/ Urban']).sum().plot(kind='bar', y=['Total disabled population - Persons','Total disabled population - Males','Total disabled population - Females'])
```

# <font color="orchid">**Insights :**</font>


*   The total 'Houseless Household' disabled persons in India is 68422.
*   In Rural india- 24913(male:14339, female: 10574)
*   In Urban india- 43509(male:29872, female:13637)





# <font color="springgreen">**Q3. What are the 'Normal Household' population in India, including rural and urban areas ?**</font>

```python
df2 = df[(df["Type of Household"]=="Normal Household") & (df["Area Name"]=="INDIA")]
df2
```

```python
df2.groupby(['Total/ Rural/ Urban']).sum().plot(kind='bar', y=['Total disabled population - Persons','Total disabled population - Males','Total disabled population - Females'])
```

*   The total 'Normal Household' disabled in india is 26441354, which is more than 'Houseless Household'
*   In rural India the "Normal Household" male population(10316977) exceeds the female population(8160289).
*   In urban India the "Normal household" male population(4438724) exceeds the female population(3525364).

# <font color="springgreen">**Q5. 'Houseless Householders' with disabilities in rural area of each state ?**</font>

```python
a = df[(df["Type of Household"]=="Houseless Household") & (df["Total/ Rural/ Urban"]=="Rural") & (df["Area Name"]!="INDIA")]
a.sort_values(by=["Total disabled population - Persons"], ascending=False)
```

# <font color="orchid">**Insights :**</font>



*   4464 persons with disabilities live in 'Houseless Household' in rural areas of Uttar Pradesh, the highest in India. Whereas no disabled person lives in 'Houseless Household' in rural areas of 'Daman & Diu', 'Lakshadweep', 'Mizoram','Chandigarh' in India
*   In Uttar Pradesh, there are more male(111) suffering from multiple disability than female(69).
